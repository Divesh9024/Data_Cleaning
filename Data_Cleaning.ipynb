```python
#importing dataset
import pandas as pd
data=pd.read_csv("food_coded.csv")
print(data)
```

```python
missing_values=data.isnull()#identify missing values
print(missing_values)
```

```python
#to calculate which column containing null values
data.isnull().sum()
```

```python
data.describe()
```

```python
data.duplicated().sum()#check for duplicates
```

```python
median=data['calories_day'].median()#to find median of calories_day
print(median)
```

```python
data['calories_day']=data['calories_day'].fillna(median)#fill null values by median of calories_day
print(data)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder_x=LabelEncoder()
data['type_sports']=label_encoder_x.fit_transform(data['type_sports'])#convert string value to integer
data
```

```python
data['type_sports']=data['type_sports'].fillna(data['type_sports'].mode()[0])
data
```

```python
data.isnull().sum()
```

```python
data['weight']=pd.to_numeric(data['weight'],errors='coerce')
data
```

```python
data['weight']=data['weight'].fillna(data['weight'].median())
print(data)
```

```python
data.isnull().sum()
```

```python
data['GPA']=pd.to_numeric(data['GPA'],errors='coerce')
data['GPA']=data['GPA'].fillna(value=data['GPA'].median())
data
```

```python
missing_values=data.isnull().sum()
print(missing_values)#all the count of isnull value is 0 it means that data is clean
```
